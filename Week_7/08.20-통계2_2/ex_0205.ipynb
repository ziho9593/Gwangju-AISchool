### 다양한 상관계수 유형

```python
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

#### 데이터 프레임을 읽어 온다:

```python
df = pd.read_csv('data_iris.csv', header='infer',encoding = 'latin1')
```

```python
df.shape
```

```python
df.head(3)
```

```python
x = df['Petal.Length']
y = df['Sepal.Length']
```

```python
plt.scatter(x, y, color= 'green', alpha=0.5, marker ='o', s=30 )
plt.xlabel('Petal Length')
plt.ylabel('Sepal Length')
plt.title('IRIS FLOWER SCATTER PLOT')
plt.show()
```

#### Pearson 상관계수

```python
np.round(st.pearsonr(x,y),3)           # 상관계수와 p-값.
```

Pandas 사용.

```python
x.corr(y)
```

```python
np.round(df.corr(),3)                  # 상관계수 행렬
```

#### Spearman 상관계수

```python
np.round(st.spearmanr(x,y),3)          # 상관계수와 p-값.
```

#### Kendall 상관계수

```python
np.round(st.kendalltau(x,y),3)         # 상관계수와 p-값.
```
