**4가지(언어, 음성, 이미지, 자율주행) 분야의 기술을 사용하는 제품 및 서비스 찾기**

## 1. 언어
- 자연어 처리
  > '자연어'란 흔히 사람들이 쓰는 일반적인 언어라고 생각하면 쉬운데, 정보 전달의 수단으로서 인간이 가지고 있는 고유의 능력입니다. 이는 인공언어에 대응되는 개념으로, 한국어나 영어, 중국어와 같이 특정한 집단에서 사용되는 모국어의 집합이라 말할 수 있습니다.<br>
  <br>
  '자연어 처리(Natural Language Processing)'는 앞서 말한 인간의 언어인 자연어를 컴퓨터를 이용해 처리하기 위한 기술 분야입니다. 이는 인간의 언어를 이해하는 머신러닝 기술을 적용해 각종 정보처리에 이용함으로써 보다 편리하고 빠르게 정보를 획득할 수 있는 장점이 있습니다. 자연어 처리는 단순 정보검색이나 질의응답 시스템 뿐만 아니라 기계로 번역하거나 자동 통역을 하고 문서 작성, 요약 분류, 철자 오류를 찾아내 수정하는 등 인가의 언어가 사용되는 많은 영역에서 응용되고 있습니다.<br>
  <br>
자연어 처리는 최근 빅데이터와 인공지능, 딥러닝의 바람을 타고 관심이 높아지고 있는 기술입니다. 대부분의 인공지능 기술처럼 이론적 성능과 기술적인 수준 사이에는 큰 차이가 있기 때문에 차이를 좁히기 위해서는 엄청난 양의 데이터와 이에 기반된 다양한 소트웨어적 노하우가 필요합니다. 이는 단순한 정의와 기존 모듈의 정의가 아니기 때문에 상당히 어렵다고 말할 수 있습니다.<br>
<br>
이러한 자연어 처리의 핵심 기술은 형태소 분석기술, 구문분석 기술, 의미분석 기술, 담화분석 기술, 단어 및 문장 생성기술 이라고 할 수 있습니다. 가장 기초적인 기술로 형태소 분석이 단어 단위에서 이루어 지고, 그 결과를 토대로 문장 단위, 문서단위의 구문분석, 의미분석, 담화분석이 진행되기 때문에 단어를 단위로 하는 분석결과를 기반으로 하는 문장, 문서 단위의 구문과 의미, 담화분석이 더욱 복잡하며 어려운 기술이라 할 수 있습니다.<br>
<br>
인공지능 연구의 궁극적 목표를 생각해보면 인간과 비슷한 수준까지 인공지능 기술을 끌어올려 다양한 개발을 하는 것인데, 이 목표를 위해서는 반드시 자연어 처리에 관한 연구가 필수적으로 이루어 져야 합니다. 아직까지도 인간수준의 자연어 처리 능력은 부족해 보이지만 최근 딥러닝(Deep learning) 으로 인해 인공지능 분야에 대한 관심도 높아지고 있으므로, 자연어 처리 기술 역시 이에 맞춰 빠르게 발전할 것으로 보입니다.

## 2. 음성
- 클로바더빙
> '클로바더빙'은 네이버의 인공지능 플랫폼 클로바(CLOVA)에서 제공하는 목소리 녹음 없이도 동영상에 AI 보이스를 입힐 수 있는 서비스입니다. 이는 텍스트 입력을 통해 합성음을 생성하고 동영상에 더빙하는 것으로 사람이 직접 목소리를 녹음하고 후속 작업하는 번거로움을 줄여 누구나 쉽게 더빙 콘텐츠를 제작할 수 있도록 합니다.<br>
<br>
클로바더빙 웹사이트에서 동영상을 추가한 후 원하는 보이스를 선택하고 내용을 입력하면 더빙 음성이 생성됩니다. 이후 사용자가 지정한 시간대에 더빙 음성을 추가하는 방식이며, 성인과 아이, 남성과 여성, 기쁨과 슬픔 등 다양한 개성을 가진 21종 보이스를 제공합니다. 이외에도 AI 보이스 속도·볼륨 조절 기능과 웃음소리, 동물소리 등 다양한 효과음을 제공해 콘텐츠 제작 편의성을 높였습니다.<br>
<br>  
클로바더빙이 제공하는 보이스는 음성합성 AI 기술인 ‘클로바 보이스’를 통해 개발한 것으로, 사용자가 입력한 문장을 자연스러운 억양과 감정으로 표현하고 노이즈 없이 선명하게 전달한다는 특장점이 있습니다. 기존 합성음 제작에는 녹음 시간이 40~100시간이 걸렸다면, 클로바 보이스는 40분 음성 데이터로 합성음을 제작할 수 있습니다.

## 3. 이미지
- CheXNeXt
  >미국 스탠퍼드대 연구팀이 개발한 것으로, 흉부 X선 이미지를 보고 14개 질환을 동시에 진단할 수 있는 인공지능 알고리즘입니다. 진단 정확도와 판독 시간 모두 방사선과 전문의를 능가한다는 평가를 받으며, 단 하나의 알고리즘만으로 수백 장의 X선 이미지를 읽을 수 있습니다.<br>
<br>
연구를 이끈 매튜 룬그렌 교수는 “책 읽는 인공지능이 문장을 읽는 것보다 훨씬 빠른 속도로 X선 이미지를 판독한다”면서 “촬영한 부위가 어디인지 알아내는 것뿐 아니라 14가지 병리학적 증상을 알아낼 수 있다”고 설명했습니다. CheXNeXt가 특별히 폐렴(Pneumonia)과 기흉(Pneumothorax) 등 14가지 병리에 대해서 전문적으로 진단할 수 있도록 훈련받았기 때문입니다.<br>
<br>
연구팀은 이 알고리즘이 얼마나 정확하게 X선 이미지를 판독하는지 알아보기 위해 방사선과 전문의가 판독한 결과와 비교하는 실험을 진행했는데, 그 결과 알고리즘이 14가지 병리학적인 특징을 찾아낼 수 있었습니다. 이 중 10가지는 방사선과 전문의보다 훨씬 정확하게 진단 내렸고, 이는 알고리즘이 전문가를 능가한 셈입니다.<br>
<br>
심지어 X선 이미지 420장을 읽는 데 방사선과 전문의는 평균 약 3시간이 걸리는데 반해, CheXNeXt는 약 90초만에 모든 이미지를 판독해 질환을 진단했습니다. 정확도뿐만 아니라 소요시간면에서도 월등히 뛰어난 결과를 보여준 것입니다.<br>
<br>
또한 인공지능은 훈련을 하면 할수록 실력이 늘기 때문에 X선 이미지를 많이 보면 볼수록 질병을 정확하게 진단할 확률도 높아지며, 지금까지 CheXNeXt가 판독한 X선 이미지는 10만 장이 넘지만 스탠퍼드대 병원에서만 촬영한 이미지라는 한계가 있다. 연구팀은 앞으로 전 세계 병원에서 촬영한 X선 이미지를 100만 장 이상 보여줄 예정이라고 말했습니다.

## 4. 자율주행
- 테슬라 오토파일럿
  >테슬라 자동차에서 사용되고 있는 ADAS(첨단 운전자 보조 시스템)로서 운전자없이 자동차를 운행하는 것을 최종 목표로 하고 있는 기술입니다. 테슬라 자동차의 배타적 경쟁력중 하나로서, 테슬라의 오토파일럿 기능을 사용하려는 목적이 테슬라 자동차의 구매결정을 하는데 큰 요소중 하나로 작용하고 있습니다.<br>
  <br>
  현재 적용된 기능은 크게 차선유지, 속도 유지, 차선 변경으로 요약되는데, 먼저 차선유지는 타 회사 제품에서도 기능을 체험해 볼 수 있으나, 급한 커브에서 조향이 안되어 차선을 이탈하거나, 부드럽지 않은 계단식 조향이나 차선내에서 좌우를 왔다 갔다하는 핑퐁식 조향을 보여 주는 것과 달리, 테슬라의 자율 주행 기능은 사람이 운전하는 것 같은 부드럽고 신뢰할 수 있는 단계로 보여집니다.<br>
  <br>
  가장 앞선 자율주행기술을 가지고 있는 것으로 알려진 웨이모가 2020년초에 1000여대의 자동차를 통해서 약 2천만마일(3천2백만km)의 실제 도로 데이터를 축적하고 있는 것으로 발표한 가운데, 테슬라는 100만대 이상의 자동차중 HW2이상의 하드웨어를 가진 80만대의 테슬라 자동차를 통해서 하루에도 웨이모의 데이터이상의 실도로 데이터를 수집할 수 있는 역량을 갖추고 있습니다. 같은 시기 테슬라는 30억마일(48억km)의 실도로 주행 데이터를 가지고 있다고 합니다.<br>
  <br>
  딥러닝에 있어서 가장 중요한 것으로 취급되는 요소는 데이터의 양으로서, MIT의 인공지능는 연구자인 Lex Fridman에 따르면 테슬라의 딥러닝이 케익이면, 웨이모의 딥러닝은 케익 위의 아이싱(장식)으로 표현했습니다. 2020년 인터뷰에 따르면 테슬라는 최근에 자율주행 프로그램의 파운데이션을 다시 작성하여 거의 완성단계에 다다른 상황이라고 합니다.
