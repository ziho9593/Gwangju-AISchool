#### 훈련 데이터와 테스트 데이터 분리

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

## 5. 모델 선택 및 학습

- 좋은 모델을 선택
- 학습 -> fit()

### KNN

```python
from sklearn.neighbors import KNeighborsRegressor

# model_knn
model_knn = KNeighborsRegressor(n_neighbors=7).fit(X_train, y_train)
```

###  Linear Regression

```python
from sklearn.linear_model import LinearRegression

# model_lr
model_lr = LinearRegression().fit(X_train, y_train)
```

### Ridge

- L2 규제 : 큰 값은 줄이고 작은 값은 크게하여 feature가 적은 경우에 숫자가 작은 feature도 학습에 참여할 수 있게 함
- alpha 값으로 설정

```python
from sklearn.linear_model import Ridge

# model_ridge
model_ridge = Ridge(alpha=0.1).fit(X_train, y_train)
```

### Lasso

- L1 규제 : feature가 많은 경우 크기가 작은 feature를 버리고 학습하는 것
- alpha 값으로 설정
- feature selection에서 사용

```python
from sklearn.linear_model import Lasso

# model_lasso
model_lasso = Lasso(alpha=0.01).fit(X_train, y_train)
```

### SVM

- 전처리로 스케일링이 필요

```python
from sklearn.svm import SVR, LinearSVR

# model_svr
model_svr = SVR(C=0.01).fit(X_train, y_train)
# model_lsvr
model_lsvr = LinearSVR(C=0.01).fit(X_train, y_train)
```

## 6. 평가

- MSE / RMSE : mean_squared_error()
-  R^2 점수 -> score(), r2_score() <br>
(정밀도(R^2) : 평균 모델과 비교하여 얼마나 유사한지를 평가)
- 교차검증

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# KNN, 선형회귀, Ridge, Lasso, SVR, LinearSVR MSE 오차점수 확인
pred = model_knn.predict(X_test)
print('KNN :', np.sqrt(mean_squared_error(pred, y_test)))

pred = model_lr.predict(X_test)
print('선형회귀 :', np.sqrt(mean_squared_error(pred, y_test)))

pred = model_ridge.predict(X_test)
print('Ridge :', np.sqrt(mean_squared_error(pred, y_test)))

pred = model_lasso.predict(X_test)
print('Lasso :', np.sqrt(mean_squared_error(pred, y_test)))

pred = model_svr.predict(X_test)
print('SVR :', np.sqrt(mean_squared_error(pred, y_test)))

pred = model_lsvr.predict(X_test)
print('LinearSVR :', np.sqrt(mean_squared_error(pred, y_test)))
```

## 7. 모델 개선

- 파라미터 튜닝 -> GridSearchCV()
- 모델을 중복해서 사용 -> Stacking, Pipeline()

#### 파라미터 튜닝

```python
import matplotlib.pyplot as plt

# KNeighborsClassifier의 n_neighbors 파라메터의 이웃 수 변화에 따른 RMSE 오차 변화
train_error = []
test_error = []

n = range(1, 15)
for i in n:
    model_knn = KNeighborsRegressor(n_neighbors=i).fit(X_train, y_train)
    
    pred = model_knn.predict(X_train)
    train_error.append( np.sqrt( mean_squared_error(pred, y_train) ) )
    
    pred = model_knn.predict(X_test)
    test_error.append( np.sqrt( mean_squared_error(pred, y_test) ) )
    
plt.plot(n, train_error)
plt.plot(n, test_error)
```

```python
# 최적의 alpha값은? (Ridge, Lasso)
import matplotlib.pyplot as plt

train_error = []
test_error = []

alpha = [0.001, 0.01, 0.1, 1, 10, 100, 1000]

for i in alpha:
    model_ridge = Ridge(alpha=i)
    model_ridge.fit(X_train, y_train)
    
    pred = model_ridge.predict(X_train)
    train_error.append(np.sqrt(mean_squared_error(pred, y_train)))
    pred = model_ridge.predict(X_test)
    test_error.append(np.sqrt(mean_squared_error(pred, y_test)))

plt.plot(alpha, train_error)  
plt.plot(alpha, test_error)  
plt.legend(["train","test"])
plt.xlim(0, 0.02)
```
