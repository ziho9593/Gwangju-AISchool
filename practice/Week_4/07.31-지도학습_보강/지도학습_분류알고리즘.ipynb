#### 훈련 데이터와 테스트 데이터 분리

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

## 5. 모델 선택 및 학습

- 좋은 모델을 선택
- 학습 -> fit()

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier

# model_knn
model_knn = KNeighborsClassifier(n_neighbors=10).fit(X_train, y_train)
```

### Ridge

```python
from sklearn.linear_model import RidgeClassifier

# model_ridge
model_ridge = RidgeClassifier(alpha=0.1).fit(X_train, y_train)
```

### Logistic Regression

- C : 규제항

```python
from sklearn.linear_model import LogisticRegression

# model_lr
model_lr = LogisticRegression(C=0.01).fit(X_train, y_train)
```

### SVM

- 전처리로 스케일링이 필요

```python
from sklearn.svm import SVC, LinearSVC

# model_svc
model_svc = SVC(C=0.01, random_state=0).fit(X_train, y_train)

# model_lsvc
model_lsvc = LinearSVC(C=0.01, random_state=0).fit(X_train, y_train)
```

## 6. 평가

- 정확도 : score()
- f1 스코어
- 교차검증

```python
# KNN, Ridge, Logistic, SVC, LinearSVC 훈련 스코어, 테스트 스코어 확인
print("KNN 훈련:", model_knn.score(X_train, y_train))
print("KNN 테스트:", model_knn.score(X_test, y_test))

print("Ridge 훈련:", model_ridge.score(X_train, y_train))
print("Ridge 테스트:", model_ridge.score(X_test, y_test))

print("Logistic 훈련:", model_lr.score(X_train, y_train))
print("Logistic 테스트:", model_lr.score(X_test, y_test))

print("SVC 훈련:", model_svc.score(X_train, y_train))
print("SVC 테스트:", model_svc.score(X_test, y_test))

print("LinearSVC 훈련:", model_lsvc.score(X_train, y_train))
print("LinearSVC 테스트:", model_lsvc.score(X_test, y_test))
```

## 7. 모델 개선

- 파라미터 튜닝 -> GridSearchCV()
- 모델을 중복해서 사용 -> Stacking, Pipeline()

#### 파라미터 튜닝

```python
import matplotlib.pyplot as plt

# KNeighborsClassifier의 n_neighbors 파라메터의 이웃 수 변화에 따른 훈련/테스트 스코어 변화
train_score = []
test_score = []

n = range(1, 15)
for i in n:
    model_knn = KNeighborsClassifier(n_neighbors=i).fit(X_train, y_train)
    train_score.append( model_knn.score(X_train, y_train) )
    test_score.append( model_knn.score(X_test, y_test) )
    
plt.plot(n, train_score)
plt.plot(n, test_score)
```
