### 0. 개요
- 프로그래밍을 배우다 보면 “알고리즘”이라는 말을 자주 듣게 됩니다. 알고리즘(Algorithm)은 컴퓨터과학에서만 쓰이는 단어는 아니며 여러 학문에서 “어떠한 문제를 해결하기 위한 일련의 절차”라는 뜻을 가지고 있습니다. 같은 문제라도 어떤 알고리즘을 적용하냐에 따라 그 효율은 크게 달라집니다. 반복되는 작업 속에서 규칙을 찾을 수 있다면 수백 번의 계산도 단 한 번의 계산으로 해결할 수도 있습니다. 수열을 공부하는 것은 이러한 관찰력과 사고력을 기르는 출발점이 됩니다.

### 1. 수열의 정의
- 수열은 이름 그대로 “숫자의 나열”을 말합니다. 
- 1, 2, 3, … 하고 자연수를 늘어놓거나, 2, 4, 6, … 처럼 짝수만 센다던가, 혹은 7, 2, 5, 9, 1, … 처럼 아무 의미가 없어보이는 수들의 나열도 모두 수열입니다. 
- 여기서의 숫자는 자연수 뿐만 아니라 음수, 실수, 복소수까지도 가능합니다.
- 1,2,3,4,5 처럼 갯수가 정해져 있어 끝이 있는 수열을 유한수열이라 부릅니다. 반면에 1,2,3,4,5,6,7,… 처럼 끝이 없이 이어지는 수열은 무한수열이라 부릅니다.
- 수열은 일반적으로 $ a_n = 1, 2, 3, 4, 5, ... $ 와 같이 표현하며, n의 범위를 보통 같이 표기합니다만 생략했다면 일반적으로 자연수라고 간주합니다. 
- 별도의 언급이 없다면 n은 항상 자연수라고 생각하면 되겠습니다.

### 2. 등차수열
- 수열 중에서도 특정한 규칙을 가지고 이어지는 수열들이 있는데, 그 중 등차수열은 수들이 일정한 간격으로 커지거나 작아지는 수열을 말합니다.
- 등차수열을 식으로 나타내면 다음과 같습니다. : $ a_n = a + (n-1)d $
- 여기서 a는 수열의 첫째항(초항이라 부릅니다)을 나타내며 d는 뒤로 갈수록 더해지는 어떤 상수, 즉 "공차"를 말합니다.
- 공차가 양수라면 뒤로 갈수록 수가 커지는 “증가수열”이 되고 음수라면 “감소수열”이 됩니다. 
- 공차가 0인 경우에는 $ a_n=a+(n−1)×0 $ 이 되기 때문에 n에 값에 상관없이 항상 같은 값을 가지는 수열이 됩니다.
- 이렇게 수열을 n에 대한 식으로 나타내면 찾고자 하는 항이 몇번째에 있던지간에 n만 집어넣으면 빠르게 그 값을 구할 수 있습니다. 
- n에 대해서 수열을 식으로 나타낸 위와 같은 식 $a_n = a + (n-1)d$ 을 수열에 대한 “일반항”이라고 부릅니다.
- 공식을 알면 수학문제를 빠르게 푸는것과 같이 수열의 일반항을 알면 수열에서 원하는 위치의 항을 빠르게 구할 수 있습니다.

<br>
<예제 - 등차수열>  

각 항마다 일정한 간격으로 더해지는 수열을 등차수열이라고 부릅니다.  

등차수열의 첫번째 항을 초항이라고 부르며 일정하게 더해지는 값은 공차라고 부릅니다.  

등차수열의 초항과 공차가 주어졌을 때, 이 등차수열의 n번째 항을 구하는 함수를 작성해 봅시다.

```python
def arithSeq(a, d, n):
    # To-do
    # 등차수열의 초항 a와 공차 d가 주어졌을 때 
    # 이 등차수열의 n번째 항을 구하는 함수를 작성하고, 이를 반환해봅시다.
    return a + (n - 1) * d
    
    
# 결과 출력을 위한 코드입니다. 자유롭게 값을 바꿔보며 확인해보세요.
print(arithSeq(1, 2, 3))
print(arithSeq(10, 5, 5))
```

### 3. 등비수열
- 등차수열이 일정한 수 만큼 “더해지는” 수열이라면 등비수열은 일정한 수 만큼 “곱해지는” 수열입니다. 
- 예를 들어 1, 2, 4, 8, 16, … 은 매 항마다 2가 곱해지는 등비수열이며, 이때 곱해지는 일정한 상수를 “등비”라고 부르고 주로 r로 표시합니다.
- 등비수열의 일반항은 $ a_n = ar^{n-1} $ 이며, a는 초항을 나타내며 r은 등비를 나타냅니다.
- 등비 수열은 등비의 범위에 따라서 다양한 모습을 보이게 됩니다. 7개의 경우로 꽤 많이 나누어지는데, 정리해보면 아래와 같습니다.
    - **r > 1**: r이 1보다 크므로 곱해질때마다 수열이 증가하게 됩니다.
    - **r = 1**: 1은 아무리 곱해도 그 값이 변하지 않으므로 제자리에 머물게 됩니다. 등차수열에서 d=0일 때와 동일합니다.
    - **0 < r < 1**: r이 곱해짐에 따라 점차 수가 작아짐을 알 수 있습니다. 계속해서 줄어들어 0에 가깝게 됩니다.
    - **r = 0**: 이 경우는 우선 넘어가겠습니다. (0의 0제곱에 대한 문제가 있기 때문입니다.)
    - **-1 < r < 0**: r이 음수이기 때문에 수열의 부호가 양수와 음수가 번갈아 나타나게 됩니다. 부호가 왔다갔다하면서 그 절대값은 계속 줄어들기 때문에 0에 가까워지는 모습을 하게됩니다. 등비의 부호가 음수인 경우에는 이렇게 값이 커졌다 줄어들었다를 반복하기 때문에 증가수열이나 감소수열이 되지 않습니다.
    - **r = -1**: 절대값은 같으면서 부호만 번갈아 나타나는 이른바 “진동수열”이 됩니다. 1, -1, 1, -1, … 처럼 2개의 값이 반복적으로 나타납니다.
    - **r < -1**: 절대값이 계속 커지면서 진동하는 수열이 됩니다.

<br>
<예제 - 등비수열>  

매 항마다 일정한 값이 곱해지는 수열을 등비수열이라고 부릅니다. 등비수열의 첫항을 초항이라부르고 곱해지는 일정한 값을 공비라고 부릅니다.  

등비수열의 초항과 공비가 주어졌을 때, 이 등비수열의 n번째 항을 구하는 함수를 작성해 봅시다.

```python
def geoSeq(a, r, n):
    # To-do
    # 등비수열의 초항 a와 공비 r가 주어졌을 때 
    # 이 등비수열의 n번째 항을 구하는 함수를 작성하고, 이를 반환해봅시다.
    return a * r ** (n - 1)


# 결과 출력을 위한 코드입니다. 자유롭게 값을 바꿔보며 확인해보세요.
print(geoSeq(1, 2, 3))
print(geoSeq(2, 4, 6))
```

### 4. 계차수열
- 등차수열이 일정한 공차가 더해지는 수열이라면, 일정한 규칙을 가진 값들이 더해지는 수열도 우리가 특별히 다루는 수열 중 하나입니다.
- 등차수열 $ b_n = 2, 4, 6, 8,... $ 이 있다고 할 때, 수열 $ a_n $ 이 초항으로 1을 가지고 "$ b_n $ 을 계차수열"로 가진다면, $ a_n = 1, 3, 7, 13, ... $ 이 됩니다.
- 이처럼 각 항 사이의 차가 새로운 수열을 이루는 경우도 있습니다. 또한, 계차수열은 등차수열 뿐만 아니라 등비수열이 될 수도 있습니다. 
- 계차수열을 가지는 수열의 일반항은 위에서 살펴본것과 같이 수열 $ a_n $ 의 n항은 초항에서부터 계차수열 $ b_n $ 을 순서대로 더하는 것으로써 표현할 수 있다.
- 따라서 $ a_n = a_1 + b_1 + b_2 + ... + b_{n-1} $ 이 되는데, 이를 일반항이라고 부르기엔 $ b_n $ 의 덧셈이 너무 거추장스럽다. 이는 다음 섹션에서 해결할 예정이다.

<br>
<예제 - 계차수열>

수열 $ a_n $ 이 초항이 $a$이고 $b_n$을 계차수열로 가지는 수열이라 할 때, 수열 $a_n$의 n번째 항을 구하는 함수를 작성해 봅시다.

```python
def seqDiff(a, b, n):
    # To-do.
    # 초항 a와 계차수열 b로 이루어진 수열 An이 주어졌을 때
    # 수열 An의 n번째 항을 구하는 함수를 작성하고, 이를 반환해봅시다.
    return a + sum(b[:n-1])
    
    
# 결과 출력을 위한 코드입니다. 자유롭게 값을 바꿔보며 확인해보세요.
print(seqDiff(1, [1, 2, 3, 4, 5], 3))
print(seqDiff(10, [2, 4, 6, 8, 10], 5))
```

### 5. 급수
- 급수는 “수열의 각 항의 합”을 일컫는 말로, 주로 $ S_n $의 형태로 나타냅니다.
- 예를 들면 등차수열 $ a_n = 2, 4, 6, 8, ... $ 에서 4항까지의 합 $ a_1 + a_2 + a_3 + a_4 $ 를 $ S_4 $ 라고 표기합니다.
- $ S_n $ 의 일반항은 수열에 따라 다릅니다만, 등차수열과 등비수열은 비교적 쉽게 나타낼 수 있습니다. (각 공식을 여기서는 다루지 않겠습니다.)

<br>
<예제 - 급수>

주어진 수열의 n항 까지의 합 $ S_n $ 를 구하는 함수를 작성해봅시다.

```python
def series(a, n):
    # To-do
    # 수열 a의 n번째 항까지의 합을 구하는 함수를 작성하고, 이를 반환해봅시다.
    return sum(a[:n])
    
    
# 결과 출력을 위한 코드입니다. 자유롭게 값을 바꿔보며 확인해보세요.
print(series([1, 2, 3, 4, 5], 5))
print(series([2, 4, 6, 8, 10], 10))
```

### 6. 점화식
- 수열 중에서는 n에 대한 일반항만으로는 나타내기 어려운 수열들도 있습니다. 등차수열이나 등비수열 외에 우리가 실제로 만나게 되는 수열들은 이처럼 규칙이 마냥 단순하지는 않습니다. 
- 경우에 따라 구하고자하는 n번째 항이 바로 앞 항인 n-1번째 항에 의해서 정해질수도 있습니다.
- 이렇게 n에 대한 식이 아닌 항들 간의 관계에 대한 식을 수열의 “점화식”이라고 부릅니다. 수열은 앞에서 배운 일반항뿐만 아니라 점화식으로도 정의할 수 있습니다.
- 일반항과 달리 점화식은 상대적으로 식이 간단하다는 장점이 있으며 일반항으로 나타내기 어려운 수열들도 표현할 수 있습니다.
- $ a_n $ 을 구하기 위해서 앞 항의 정보가 필요하기 때문에 $ a_1 $ 부터 차근차근 계산해서 올라가야 한다는 단점이 있지만, 이러한 반복 작업은 컴퓨터의 전분 분야이므로 컴퓨터에게 맡기면 됩니다.

<br>
<예제 - 점화식>  

n에 의존하지 않고 항들 간의 관계로 수열을 나타내는 식을 점화식이라고 부릅니다.  

초항 a와 상수 p, q가 주어졌을 때 다음의 점화식을 만족하는 수열의 n번째 항을 구하는 함수를 작성해 봅시다.

> $ a_1 = a$, $ a_{n+1} = pa_n + q $

```python
def recurrSeq(a, p, q, n):
    # To-do
    # 주어진 점화식(p*a + q)을 만족하는 수열의 n번째 항을 구하고, 이를 반환해봅시다.
    for _ in range(n-1):
        a = p * a + q
    return a
    
    
# 결과 출력을 위한 코드입니다. 자유롭게 값을 바꿔보며 확인해보세요.
print(recurrSeq(1, 2, 3, 5))
print(recurrSeq(10, 3, 5, 10))
```

### 7. 수학적 귀납법
- 수학적 귀납법은 어떠한 등식이 모든 자연수 n에 대해 성립함을 보이는데 주로 쓰이는 증명법입니다. 기본적인 과정은 다음과 같습니다.
    1. 주어진 등식이 n=1일 때 식이 성립함을 보인다.
    2. 주어진 등식이 n=k일 때 성립한다고 가정했을 때 n=k+1에서도 성립함을 보인다.
    3. n=1일 때가 성립하므로 (2)에 의해 n=2일때 역시 성립한다.
    4. n=2일 때가 성립하므로 (2)에 의해 n=3일때 역시 성립한다.
    5. n=3일 때가 성립하므로 (2)에 의해 n=4일때 역시 성립한다.  
       …
    6. 따라서 모든 자연수 n에 대해 식이 성립한다.
- 이렇듯 맨 첫 번째 n에 대해 성립하고 앞 항이 성립하면 뒷 항 역시 성립함을 증명하여 도미노가 쓰러지듯이 연쇄 반응을 통해 증명하는 방법입니다.

### 8. 분할정복
- 수학적 귀납법의 연쇄반응을 이용한 증명법은 프로그래밍의 알고리즘에서도 사용될 수 있습니다. 그 중 대표적인 알고리즘이 “분할정복” 알고리즘입니다.
- 분할정복 알고리즘은 하나의 큰 문제를 여러 개의 풀기 쉬운 작은 문제로 나누어 해결하는 방법입니다.
- 큰 문제를 다루는 상위 함수는 작은 문제를 다루는 하위 함수를 호출하는 구조로 설계되며 하위 함수가 잘 작동한다고 가정하고 상위 함수를 설계합니다.
- n = k일때 식이 성립한다고 가정할 때 n = k+1일 때도 식이 성립함을 보이는 귀납법과 같은 맥락입니다.

<br>
<예제 - 분할정복>  

분할정복법은 바로 해결하기 어려운 하나의 큰 문제를 비교적 해결하기 쉬운 여러 개의 문제로 쪼개어 해결하는 문제입니다. 

재귀함수를 이용하여 쉽게 해결할 수 있을 때까지 문제를 쪼개게 됩니다.  

분할정복법을 사용하여 리스트를 오름차순으로 정렬하는 알고리즘을 완성해 봅시다.

```python
def divSort(a):
    # 리스트의 길이가 1보다 작거나 같은 경우
    if len(a) <= 1:
        return a
    # 리스트의 길이가 2인 경우
    elif len(a) == 2:
        # To-do: 정렬된 리스트를 반환한다
        if a[0] > a[1]:
            return [a[1], a[0]]
        else:
            return a
    
    
    # 리스트의 길이가 2보다 큰 경우
    else:
        l = len(a)
        # 기준이 되는 요소를 하나 정한다
        m = a[l//2]
        d1 = [i for i in a if i < m]
        d2 = [i for i in a if i > m]
        
        # To-do: m을 기준으로 2개의 리스트를 나눈다
        s1 = divSort(d1)
        s2 = divSort(d2)
        return s1 + [m] + s2
        

# 결과 출력을 위한 코드입니다. 자유롭게 값을 바꿔보며 확인해보세요.
print(divSort([2, 3, 1]))
print(divSort([7, 9, 1, 10, 19]))
```

- <문제 1 - 이자 계산>  
  수열 문제 중에서 실생활과 가장 연관이 있는 예 중 하나는 이자 계산입니다. 이자가 붙는 방식에는 단리와 복리가 있는데, 이 중 복리는 이자 계산 시 (원금+이자)에 이자가 붙게 되어 이자가 빠른 속도로 늘어납니다.  

  가령 원금이 100만원이고 연 이자율이 1%라면 1년 후에는 원금 100만원에 이자 1만원, 2년 후에는 101만원에 이자가 1.01만원이 붙어 총 이자는 1 + 1.01 = 2.01만원이 됩니다.

  원금 a와 연 이자율 r이 주어졌을 때, n년 후의 총 이자 금액을 계산하는 함수를 작성해 봅시다.  

```python
def interest(a, r, n):
    # To-do
    # 등비수열을 통해 n년 후의 금액이 어떻게 되는지 계산해봅시다.(원금 * (1+ 연이자율)^n)
    # 총 금액에서 원금을 뺀 금액(이자)을 계산한 후, 이를 반환해보세요.
    result = 0
    for _ in range(n):
        result += a * r
        a = a + a * r
        
    return result
    
    
# 결과 출력을 위한 코드입니다. 자유롭게 값을 바꿔보며 확인해보세요.
print(interest(1000000, 0.01, 2))
print(interest(1000000, 0.02, 3))
print(interest(1000000, 0.03, 4))
```

- <문제 2 - 재귀함수>  

  수열에서의 점화식은 프로그래밍에 있어 재귀함수를 작성하는 것과 비슷합니다. 함수 안에서 또 다시 함수를 부르는 모양이 마치 이전 항을 사용하여 다음 항을 구하는 것과 같습니다.  

  점화식 문제를 for문을 사용하지 않고 재귀함수를 사용하여 풀어 봅시다.  

  $ a_1 = a $, $ a_{n+1} = pa_n + q$

```python
def recurrSeq2(a, p, q, n):
    # To-do
    # [실습5]와 달리 반복문을 사용하지 않고 재귀함수를 사용하여 
    # 점화식을 구현하는 코드를 작성해보세요!
    if n == 1:  
        return a
    else:
        a = p*a + q
        return(recurrSeq2(a, p, q, n-1))
    
    
# 결과 출력을 위한 코드입니다. 자유롭게 값을 바꿔보며 확인해보세요.
print(recurrSeq2(1, 2, 3, 5))
print(recurrSeq2(3, 5, 6, 5))
```

- <문제 3 - 세제곱의 합>  

  다음과 같이 정의된 수열 $a_n$ 의 n항까지의 합 $S_n$을 구하는 함수를 작성해 봅시다.  

  $a_n=1,8,27,64,...,k^3,...$  
  
  $S_n=\sum_{k-1}^{n}a_n$

```python
def cubeSum(n):
    # To-do
    # 1부터 n까지의 세제곱을 모두 합하는 함수를 작성하고, 이를 반환해보세요.
    result = 0
    for i in range(n+1):
        result += i**3
    return result
    

# 결과 출력을 위한 코드입니다. 자유롭게 값을 바꿔보며 확인해보세요.
print(cubeSum(3))
print(cubeSum(5))
```

- <문제 5 - 피보나치 수 2>  

  어떤 수 n이 주어졌을 때 n을 넘는 가장 작은 피보나치 수열은 몇 번째 항인지 구하는 함수를 작성해 봅시다.

```python
def fibonacci2(n):
    # To-do
    # [미션4]에서 작성한 함수를 활용해서 n을 넘는 가장 작은 피보나치 수열의 항 번호를 찾아 반환해보세요.
    temp_1 = 1
    temp_2 = 0
    result = 1
        
    for i in range(1, n+2):
        temp_1 = temp_2
        temp_2 = result
        result = temp_1 + temp_2
    
        if result > n:
            return i+1
    
    
# 결과 출력을 위한 코드입니다. 자유롭게 값을 바꿔보며 확인해보세요.
print(fibonacci2(1))
print(fibonacci2(20))
```
