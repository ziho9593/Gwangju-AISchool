```python
import json

file_path = './students.json'

def load_data():
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
    return data

def save_data(data):
    with open(file_path, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)
```

```python
from getpass import getpass

def validate_input(input_text):
    if len(input_text) > 12:
        return 'validate_err'
    else:
        validated_input = input_text.lower()
        return validated_input

def get_user_input(input_type):
    is_validate = False
    input_count = 0
    result = ''

    while is_validate is False and input_count < 4:
        if input_count == 3:
            print('규칙을 3번 틀리셨습니다! 프로그램이 종료됩니다.')
            exit()

        if input_count != 0:
            print(f'{input_count}번 틀리셨습니다. 규칙을 확인해주세요')

        print(f'{input_type} - 12자 이하, 소문자만 가능')

        if input_type == '아이디':
            raw_input = input(f'* {input_type}를 입력하세요 : ')
        else:
            raw_input = getpass()

        user_input = validate_input(raw_input)

        if user_input == 'validate_err':
            input_count += 1
        else:
            is_validate = True
            input_count = 0
            result = user_input
            break
            
    return result
```

```python
def signup():

    user_schema = {
        'id': '',
        'password': '',
        'name': '',
        'promise': '',
        'interest': ''
    }

    # user_schema 로 만들기
    user_schema['id'] = get_user_input('아이디')
    students_data = load_data()

    for student in students_data['students']:
        # 아이디 존재하는지 찾기
        if student['id'] == user_schema['id']:
            print('\n ********** 같은 아이디가 존재합니다! ********** \n')
            return 'same_id_exist'

    user_schema['password'] = get_user_input('비밀번호')
    user_schema['name'] = input('이름을 입력해주세요 : \n')
    user_schema['promise'] = input('5개월 간의 다짐을 입력해주세요 : \n')
    user_schema['interest'] = input('관심분야를 입력해주세요 : \n')

    # json 파일에 write
    students_data['students'].append(user_schema)

    save_data(students_data)

    print('\n회원가입이 완료되었습니다! 로그인 후 사용해주세요.\n')

    # 회원가입이 완료되었다면 로그인 진행
    run_login()
```

```python
def login():
    students_data = load_data()

    login_user_input = {
        'id': '',
        'password': '',
    }

    login_user_input['id'] = get_user_input('아이디')
    login_user_input['password'] = get_user_input('비밀번호')
    students_data['students']

    for student in students_data['students']:
        # 아이디 존재하는지 찾기
        if student['id'] == login_user_input['id']:
            if student['password'] == login_user_input['password']:
                user_name = student['name']
                print(f'\n {user_name}님, 로그인에 성공하셨습니다.!')
                return student

            # 비밀번호 틀린 경우
            else:
                print('비밀번호를 확인해주세요!')
                return 'wrong_password'

        elif students_data['students'][-1]['id'] == student['id']:
            print('존재하지 않는 아이디 입니다 !')
            return 'no_exist_id'

    return
```

```python
def run_login():

    # 로그인 과정 실행
    login_count = 0

    while True and login_count < 4:
        if login_count == 3:
            print('규칙을 3번 틀리셨습니다! 보안을 위해 프로그램이 종료됩니다.')
            exit()

        login_count += 1
        
        login_student = login()

        if login_student == 'wrong_password' or login_student == 'no_exist_id':
            continue
        else:
            break
            
    after_authenticated(login_student)
```

```python
def after_authenticated(login_student):

    # 로그인 성공
    which_page = input('1 - 마이페이지 보기 / 2 - 다른 유저 보기 / 3 - 로그아웃 : ')

    if (which_page == '1'):
        print(login_student)
    elif (which_page == '3'):
        exit()
    else:
        students_data = load_data()

        student_order = 1

        students_key_order = {}

        print('디테일을 보고싶은 학생의 번호를 입력하세요')

        for student in students_data['students']:
            student_name = student['name']
            print(f'{student_order} - {student_name}')
            students_key_order[student_order] = student
            student_order += 1
            print('------------------------------------')

        target_key_order = int(input("번호 : "))

        print(students_key_order[target_key_order])

    exit()
```

```python
def main():
    while True:
        # 1 - 회원가입 / 2 - 로그인
        signup_or_login = input('1 - 회원가입 / 2 - 로그인 : ')

        if signup_or_login == '1':
            # 회원가입 과정 실행
            signup()
        elif signup_or_login == '2':
            run = run_login()
        else:
            print('잘못 입력하셨습니다!')
```

```python
main()
```
