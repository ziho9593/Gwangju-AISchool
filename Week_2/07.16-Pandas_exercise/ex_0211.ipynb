### 멀티 인덱스 데이터 프레임

```python
import pandas as pd
import numpy as np
# import os
```

#### 신규 생성:

```python
my_header = ['a','b','c']
```

```python
my_index_out = ['G1']*3 + ['G2']*3
my_index_in = [1,2,3]*2
```

```python
my_index_zipped = list(zip(my_index_out, my_index_in))
my_index_zipped
```

```python
my_index = pd.MultiIndex.from_tuples(my_index_zipped) # 튜플을 멀티 인덱스로 받아 저장
# my_index.names = ["big", "small"]
my_index
```

```python
df = pd.DataFrame(data=np.random.randn(6,3),index=my_index,columns=my_header)
df
```

```python
my_index = pd.MultiIndex.from_tuples(my_index_zipped) 
my_index.names = ["big", "small"] # 멀티 인덱스의 이름을 설정할 수 있음
my_index
```

```python
df = pd.DataFrame(data=np.random.randn(6,3),index=my_index,columns=my_header)
df
```

### 피보팅

```python
df.groupby('gender').height.mean() 
```

```python
pd.pivot_table(df, index=['gender'], values=['height']) 
```

```python
df.groupby(['gender', 'absence']).height.mean() 
```

```python
pd.pivot_table(df, index=['gender', 'absence'], values=['height']) 
```

```python
pd.pivot_table(df, index=['gender'], columns=['absence'], values=['height']) # 그루핑과 다르게 인덱스를 열로 지정할 수 있음
```

```python
pd.pivot_table(df, index=['gender'], columns=['absence', 'bloodtype'], values=['height']) 
```

```python
pd.pivot_table(df, index=['gender', 'bloodtype'], columns=['absence'], values=['height']) 
```

```python
pd.pivot_table(df, index=['gender', 'bloodtype'], columns=['absence'], values=['height'], fill_value=0) 
```

```python
df.groupby(['gender', 'absence'])[['height', 'weight']].std() 
```

```python
pd.pivot_table(df, index=['gender', 'absence'], values=['height', 'weight'], fill_value=0, aggfunc=np.std) # aggfunc로 계산할 값 지정
```

```python
# aggfunc를 딕셔너리 형태로 넣어 각자 다른 값을 구할 수 있음 
pd.pivot_table(df, index=['gender', 'absence'], values=['height', 'weight'], fill_value=0, aggfunc={'height': np.std, 'weight': np.mean}) 
```

```python
df = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo",
                          "bar", "bar", "bar", "bar"],
                    "B": ["one", "one", "one", "two", "two",
                          "one", "one", "two", "two"],
                    "C": ["small", "large", "large", "small",
                          "small", "large", "small", "small",
                          "large"],
                    "D": [1, 2, 2, 3, 3, 4, 5, 6, 7],
                    "E": [2, 4, 5, 5, 6, 6, 8, 9, 9]})
df
```

```python
# A,B의 값으로 인덱스, C의 값으로 컬럼, 실제 셀에 들어가는 값은 E의 평균.
dfr = pd.pivot_table(df, index=['A','B'], columns='C', values='E')      
dfr
```

```python
dfr.columns
```

```python
dfr.index # 행의 인덱스가 멀티인덱스 !
```

```python
pd.pivot_table(df, index=['A','B'], columns='C', values='E', fill_value=0)
```

```python
pd.pivot_table(df, index=['A','B'], columns='C', values='E', aggfunc=np.mean, fill_value=0)
```

```python
# C의 값으로 인덱스, A, B의 값으로 컬럼, 실제 셀에 들어가는 값은 E의 평균.
dfr = pd.pivot_table(df, index='C', columns=['A','B'], values='E')
dfr
```

```python
dfr.index
```

```python
dfr.columns # 이제는 컬럼이 멀티인덱스 !
```

```python
# A,B의 값으로 인덱스, C의 값으로 컬럼, 실제 셀에 들어가는 값은 E의 중양값.
pd.pivot_table(df, index=['A','B'], columns='C', values='E', aggfunc=np.median,fill_value=0)
```

```python
pd.pivot_table(df, index=['A','B'], columns='C', values='E', aggfunc=np.sum, fill_value=0)
```

```python
# D와 E의 그룹평균.
pd.pivot_table(df, index=['A','B'], values=['D','E'], aggfunc=np.mean)
```

```python
# 위 결과와 비교.
df.groupby(['A','B'])[['D','E']].mean()
```

```python
# D, E를 다른 방법으로 집계함.
pd.pivot_table(df, index=['A','B'], values=['D','E'], aggfunc={'D':np.mean,'E':np.median})
```
